```python
def decode_string(string):
    decode = []

    def char_decoder(string, index, decode):
        if index > len(string) - 1:
            return decode
        
        look_ahead = re.findall('\(\d+x\d+\)', string[index: index+5])
        if  len(look_ahead) > 0:
            chars_num = int(look_ahead[0][1])
            copy_num = int(look_ahead[0][-2])
            chars = string[index+5:index+5+chars_num] * copy_num
            index = index + 5 + chars_num
        else:
            chars = string[index]
            index += 1
        decode += chars

        return char_decoder(string, index, decode)
    
    decoded_str = char_decoder(string, 0, decode)
    
    return ''.join(decoded_str)
```

```python
def decode_string(string):
    decode = []

    def char_decoder(string, index, decode):

        if index > len(string) - 1:
            return decode
        
        if string[index] == '(':
            match = re.search('\(\d+x\d+\)', string[index:])
            if match:
                match_len = len(match.group())
                values = re.findall('\d+', match.group())
                chars_num = int(values[0])
                copy_num = int(values[1])
                chars = string[index+ match_len:index+match_len+chars_num] * copy_num
                index = index + match_len + chars_num
        else:
            chars = string[index]
            index += 1
        decode += chars

        return char_decoder(string, index, decode)
    
    decoded_str = char_decoder(string, 0, decode)
    
    return ''.join(decoded_str)
```
