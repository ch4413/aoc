```python
dims = text.shape

count = 0
for i in range(dims[0]):
    for j in range(dims[1]):
        #forward
        if j+4 < dims[0]:
            if ''.join(text[i, j:j+4]) == 'XMAS':
                count += 1
                print('forward', i, j)
        # Backward
        if j-3 > 0:
            if ''.join(text[i, j-3:j+1]) == 'SAMX':
                count += 1
                print('backward', i, j)
        # # Updward
        if i-3 > 0:
            if ''.join(text[i-3:i+1, j ]) == 'SAMX':
                count += 1
                print('upward', i, j)
        # # # Downward
        if i+4 < dims[1]:
            if ''.join(text[i:i+4, j]) == 'XMAS':
                count += 1
                print('downard', i, j)
        # # # Diagonals
        diagonals = [
            [(i + k, j + k) for k in range(4)],
            [(i - k, j + k) for k in range(4)],
            [(i + k, j - k) for k in range(4)],
            [(i - k, j - k) for k in range(4)]
                    ]
        if all(0 <= point[0] < dims[0] and 0 <= point[1] < dims[1] for point in diagonals[0]):
            if ''.join([text[k] for k in diagonals[0]]) == 'XMAS':
                print('diag_dr', i, j)
                count += 1
        if all(0 <= point[0] < dims[0] and 0 <= point[1] < dims[1] for point in diagonals[1]):
            if ''.join([text[k] for k in diagonals[1]]) == 'XMAS':
                print('diag_dl', i, j)
                count += 1
        if all(0 <= point[0] < dims[0] and 0 <= point[1] < dims[1] for point in diagonals[2]):
            if ''.join([text[k] for k in diagonals[2]]) == 'XMAS':
                count += 1
                print('diag_ur', i, j)
        if all(0 <= point[0] < dims[0] and 0 <= point[1] < dims[1] for point in diagonals[3]):
            if ''.join([text[k] for k in diagonals[3]]) == 'XMAS':
                count += 1
                print('diag_ul', i, j)
count
```
