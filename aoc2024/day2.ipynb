```python
text = '''7 6 4 2 1
1 2 7 8 9
9 7 6 2 1
1 3 2 4 5
8 6 4 4 1
1 3 6 7 9'''.strip().split('\n')
```

```python
count = 0
for i in text:
    num_list = list(map(int, i.split(' ')))
    diffs = [num_list[i+1] - num_list[i] for i in range(len(num_list) - 1)]
    if (all([0 < x  for x in diffs]) | all([0 > x  for x in diffs])) & (all([0 < abs(x) < 4 for x in diffs])):
        count += 1
        print(num_list)
count
```

```python
def indices_of_zero(lst):
    return [i for i, value in enumerate(lst) if value == 0]

count = 0
for i in text:
    num_list = list(map(int, i.split(' ')))
    diffs = [num_list[i+1] - num_list[i] for i in range(len(num_list) - 1)]
    if (all([0 < x  for x in diffs]) | all([0 > x  for x in diffs])) & (all([0 < abs(x) < 4 for x in diffs])):
        count += 1
        print(num_list)
    else:
        count_changes = 0
        for i in range(len(num_list)):
            new_list = num_list[:i] + num_list[i+1:]
            diffs = [new_list[i+1] - new_list[i] for i in range(len(new_list) - 1)]
            if (all([0 < x  for x in diffs]) | all([0 > x  for x in diffs])) & (all([0 < abs(x) < 4 for x in diffs])):
                count_changes += 1
        if count_changes > 0:
            count += 1
            print(num_list)
count
```
