```python
import numpy as np
import re
```

```python
with open('input13.txt', "r") as file:
        text = file.read().strip().split('\n')

# text = '''Button A: X+94, Y+34
# Button B: X+22, Y+67
# Prize: X=10000000008400, Y=10000000005400

# Button A: X+26, Y+66
# Button B: X+67, Y+21
# Prize: X=10000000012748, Y=10000000012176

# Button A: X+17, Y+86
# Button B: X+84, Y+37
# Prize: X=10000000007870, Y=10000000006450

# Button A: X+69, Y+23
# Button B: X+27, Y+71
# Prize: X=10000000018641, Y=10000000010279'''.strip().split('\n')
```

```python
all_mins = []
for i in range(0, len(text), 4):
    values = [np.array(re.findall('\d+', x)).astype('int') for x in text[i:i+3]]
    results = []
    for i in range(100):
        for j in range(100):
            if all(values[0] * i + values[1] * j == values[2]):
                results.append(i * 3 + j * 1)
    if results != []:
        all_mins.append(min(results))   
sum(all_mins)
```

```python
tot = 0
for i in range(0, len(text), 4):
    values = [np.array(re.findall('\d+', x)).astype('float') for x in text[i:i+3]]
    a1, b1, c1 = values[0][0], values[1][0], values[2][0] + 10000000000000
    a2, b2, c2 = values[0][1], values[1][1], values[2][1] + 10000000000000
    factor = a2 / a1
    new_b = b2 - factor * b1
    new_c = c2 - factor * c1
    
    y = new_c / new_b
    x = (c1 - b1 * y) / a1
    if (abs(y - round(y)) + abs(x - round(x))) < 0.001:
        tot += x * 3 + y * 1
        print(x, y)
tot
```
