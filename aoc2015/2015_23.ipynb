```python
import re
```

```python
with open('input23.txt', "r") as file:
        text = file.read().strip().split('\n')
```

```python
def read_commands(text, dict_values):
    def update_ab(text, dict_values):
        index = dict_values['index']
        if index > len(text) - 1:
            print(f"Final Values: b is {dict_values['b']}, a is {dict_values['a']}")
            return
        print(f"b is {dict_values['b']}, a is {dict_values['a']} and the next command is {text[index]}")
        if text[index][:3] == 'hlf':
            dict_values[text[index][4]] = dict_values[text[index][4]] / 2
            dict_values['index'] += 1
        if text[index][:3] == 'tpl':
            dict_values[text[index][4]] = dict_values[text[index][4]] * 3
            dict_values['index'] += 1
        if text[index][:3] == 'inc':
            dict_values[text[index][4]] += 1
            dict_values['index'] += 1
        if text[index][:3] == 'jmp':
            value = int(re.findall(r'[-+]\d+', text[index])[0])
            dict_values['index'] += value
        if text[index][:3] == 'jie':
            if (dict_values[text[index][4]] % 2) == 0:
                value = int(re.findall(r'[-+]\d+', text[index])[0])
                dict_values['index'] += value
            else:
                dict_values['index'] += 1
        if text[index][:3] == 'jio':
            if dict_values[text[index][4]] == 1:
                value = int(re.findall(r'[-+]\d+', text[index])[0])
                dict_values['index'] += value
            else:
                dict_values['index'] += 1
        return update_ab(text, dict_values)
    update_ab(text, dict_values)
```

```python
dict_values = {'a': 0, 'b': 0, 'index': 0}
read_commands(text, dict_values)
```
